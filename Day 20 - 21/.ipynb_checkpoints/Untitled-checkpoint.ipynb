```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
data = load_iris()

df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df
```

```python
 kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
```

```python
df['Cluster'] = kmeans.fit_predict(df)
```

```python
df
```

```python
# Visualize clusters
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['Cluster'], cmap='viridis')
```

```python
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title("K-Means Clustering on Iris Dataset")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data  # features
y = iris.target  # target labels
```

```python
Z = linkage(X, method='ward')
```

```python
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=iris.target_names[y], leaf_rotation=90, leaf_font_size=12)

plt.title('Hierarchical Clustering Dendrogram of Iris Dataset')
plt.xlabel('Iris Species')
plt.ylabel('Distance')
plt.axhline(y=5, color='r', linestyle='--')  # Optional line to indicate cut-off
plt.show()
```

```python
reduced_labels = [
iris.target_names[label] if index % 10 == 0 else '' 
    for index, label in enumerate(y)
]
```

```python
plt.figure(figsize=(10, 7))
 # Create the dendrogram with the reduced labels
dendrogram(Z, labels=reduced_labels, leaf_rotation=90, leaf_font_size=10)
 
plt.title('Hierarchical Clustering Dendrogram of Iris Dataset')
plt.xlabel('Iris Species')
plt.ylabel('Distance')
plt.axhline(y=7, color='g', linestyle='-.')  # Optional line to indicate cut-off
plt.axvline(x=1000, color='r', linestyle='-.')  # just fun

plt.show()
```
